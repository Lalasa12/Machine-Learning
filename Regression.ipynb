# Linear Regression

```python
from sklearn.linear_model import LinearRegression
X = [[1],[3],[5],[7]]
y = [30000,50000,70000,80000]
model = LinearRegression()
model.fit(X,y)
prediction=model.predict([[9]])
print(prediction)
```

# Multiple Linear Regression

```python
from sklearn.linear_model import LinearRegression
X = [[1,1,1],[2,2,2],[3,3,3],[4,4,4]]
y = [30000,50000,70000,80000]
model = LinearRegression()
model.fit(X,y)
prediction=model.predict([[5,5,5]])
print(prediction)
```
